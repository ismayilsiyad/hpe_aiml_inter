# Speech Recognition

For speech recognition we are using the module named speech_recognition.  It is library for performing speech recognition, with support for several engines and APIs, online and offline.

```python
!pip install SpeechRecognition
```

```python
import speech_recognition as sr
#https://realpython.com/python-speech-recognition/
r = sr.Recognizer()
with sr.Microphone() as source:
    print("Speak Anything (speak for atleast one minute for more entries in word cloud) :")
    audio = r.listen(source)
    try:
        text = r.recognize_google(audio)
        print("You said : {}".format(text))
    except:
        print("Sorry could not recognize what you said")
```

# Speech Synthesis

pyttsx3 is a text-to-speech conversion library in Python. Unlike alternative libraries, it works offline and is compatible with both Python 2 and 3. it is a very easy to use tool which converts the entered text into speech.

```python
!pip install pyttsx3
```

 we are invoking the pyttsx3.init() factory function to get a reference to a pyttsx3. engine instance

```python
#text to speech library
import pyttsx3

synthesizer = pyttsx3.init()
```

Now we can try the conversion of a static text to speech

```python
synthesizer.say("Hello World") 
synthesizer.runAndWait() 
synthesizer.stop()
```

We can convert a dynamic user text to speech

```python
#convert text to speech
my_text = input("Enter the text: ")
synthesizer.say(my_text) 
synthesizer.runAndWait() 
synthesizer.stop()
```

Sometimes it is better to convert user text to an audio file

```python
#save the audio file converted from text
synthesizer.save_to_file("Hello World", 'test.mp3')
synthesizer.runAndWait()
```

You can search the test.mp3 file in the working directory and use any mp3 player to play the audiofile. Even we can convert a dynamic user text to audiofile

```python
my_text = input("Enter the text: ")
synthesizer.save_to_file(my_text, 'test1.mp3')
synthesizer.runAndWait()
```

We can customize the properties of generated voices. By default it is a male voice. we can chanege that to a female voice. Let's check what are all the available voice options

```python
#get property of the saved mp3
voices = synthesizer.getProperty('voices')
for voice in voices: 
    print("Voice:") 
    print("ID: %s" %voice.id) 
    print("Name: %s" %voice.name) 
    print("Age: %s" %voice.age) 
    print("Gender: %s" %voice.gender) 
    print("Languages Known: %s" %voice.languages)
```

We can change the voice to zira from default

```python
synthesizer.setProperty('voice', voices[1].id)
```

```python
synthesizer.say("Hello World") 
synthesizer.runAndWait() 
synthesizer.stop()
```

Now we can change the rate of speech to 100 

```python
synthesizer.setProperty('rate', 100)
```

```python
synthesizer.say("Hello World") 
synthesizer.runAndWait() 
synthesizer.stop()
```

Now we can change the volume to 0.5

```python
synthesizer.setProperty('volume', 0.5)
```

```python
synthesizer.say("Hello World") 
synthesizer.runAndWait() 
synthesizer.stop()
```
