# Speech Recognition

For speech recognition we are using the module named speech_recognition.  It is library for performing speech recognition, with support for several engines and APIs, online and offline.

```python
%pip install SpeechRecognition==3.8 pydub
```

```python
import speech_recognition as sr
r = sr.Recognizer()
```

```python
sr.__version__
```

```python
filename = "test.wav"
```

```python
# open the file
with sr.AudioFile(filename) as source:
    # listen for the data (load audio to memory)
    audio_data = r.record(source)
    # recognize (convert from speech to text)
    text = r.recognize_google(audio_data)
    print(text)
```

# Word Cloud from recognized text

we are using the library wordcloud for plotting the wordcloud from the recognized text

```python
!pip install wordcloud
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
#Count the words collected from audio on Cloud
class WordCloudGeneration:

    def create_word_cloud(self, final_data):
        # initiate WordCloud object with parameters width, height, maximum font size and background color
        # call the generate method of WordCloud class to generate an image
        wordcloud = WordCloud(width=1600, height=800, max_words=10, max_font_size=200, background_color="black").generate(final_data)
        # plt the image generated by WordCloud class
        plt.figure(figsize=(12,10))
        plt.imshow(wordcloud)
        plt.axis("off")
        plt.show()

wordcloud_generator = WordCloudGeneration()
wordcloud_generator.create_word_cloud(text)
```

# Speech Synthesis

pyttsx3 is a text-to-speech conversion library in Python. Unlike alternative libraries, it works offline and is compatible with both Python 2 and 3. it is a very easy to use tool which converts the entered text into speech.

```python
%pip install pyttsx3
```

 we are invoking the pyttsx3.init() factory function to get a reference to a pyttsx3. engine instance

```python
!lspci -v | grep Audio
```

```python
#text to speech library
import pyttsx3

synthesizer = pyttsx3.init()
```

Now we can try the conversion of a static text to speech

```python
synthesizer.say("India is my country and all Indians are my brothers and sisters. I love my country and I am proud of its rich and varied heritage. I shall always strive to be worthy of it. I shall give respect to my parents, teachers and elders and treat everyone with courtesy. To my country and my people, I pledge my devotion. In their well being and prosperitys alone, lies my happiness.") 
synthesizer.runAndWait() 
synthesizer.stop()
```

We can convert a dynamic user text to speech

```python
#convert text to speech
my_text = input("Enter the text: ")
synthesizer.say(my_text) 
synthesizer.runAndWait() 
synthesizer.stop()
```

Sometimes it is better to convert user text to an audio file

```python
#save the audio file converted from text
synthesizer.save_to_file("India is my country and all Indians are my brothers and sisters. I love my country and I am proud of its rich and varied heritage. I shall always strive to be worthy of it. I shall give respect to my parents, teachers and elders and treat everyone with courtesy. To my country and my people, I pledge my devotion. In their well being and prosperitys alone, lies my happiness.", 'test.mp3')
synthesizer.runAndWait()
```

You can search the test.mp3 file in the working directory and use any mp3 player to play the audiofile. Even we can convert a dynamic user text to audiofile

```python
my_text = input("Enter the text: ")
synthesizer.save_to_file(my_text, 'test1.mp3')
synthesizer.runAndWait()
```

We can customize the properties of generated voices. By default it is a male voice. we can chanege that to a female voice. Let's check what are all the available voice options

```python
#get property of the saved mp3
voices = synthesizer.getProperty('voices')
for voice in voices: 
    print("Voice:") 
    print("ID: %s" %voice.id) 
    print("Name: %s" %voice.name) 
    print("Age: %s" %voice.age) 
    print("Gender: %s" %voice.gender) 
    print("Languages Known: %s" %voice.languages)
```

We can change the voice to zira from default

```python
synthesizer.setProperty('voice', voices[1].id)
```

```python
synthesizer.say("India is my country and all Indians are my brothers and sisters. I love my country and I am proud of its rich and varied heritage. I shall always strive to be worthy of it. I shall give respect to my parents, teachers and elders and treat everyone with courtesy. To my country and my people, I pledge my devotion. In their well being and prosperitys alone, lies my happiness.") 
synthesizer.runAndWait() 
synthesizer.stop()
```

Now we can change the rate of speech to 100 

```python
synthesizer.setProperty('rate', 100)
```
